```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

### Line plotting

```python
# Generate arrays for x, y, and z
x = np.arange(1, 11)
y = x * x
z = x * (x + 5)

# Create a plot for y and z
plt.figure(figsize=(8, 6))  #specify the figure size for better visualization
plt.title('Function Comparison')
plt.xlabel("x")
plt.ylabel("y & z")

# Plot y and z lines, add labels, and a legend
plt.plot(x, y, label="Y")
plt.plot(x, z, label="Z")
plt.legend()

plt.show() #not needed always but stops extra info from showing
```

### Bar plotting

```python
# Create a bar plot for y
plt.figure(figsize=(8, 6))  #specify the figure size for better visualization
plt.title('Bar Plot for Y')
plt.xlabel("x")
plt.ylabel("y")
plt.bar(x, y)

plt.show() #not needed always but stops extra info from showing
```

```python
# Read data from the CSV file into a DataFrame
sample_data = pd.read_csv("sample_data.csv")
```

```python
# Display the first few rows of the DataFrame to get a quick overview of the data
sample_data.head()
```

```python
# Set the title and labels for the plot
plt.title("Scatter Plot of A against B & C")
plt.xlabel('A')
plt.ylabel('B & C')

# Plot the data for column C with red circles and label
plt.plot(sample_data['column_a'], sample_data['column_c'], 'o', color="red", label="C")

# Plot the data for column B with yellow 'x' markers and label
plt.plot(
    sample_data['column_a'],
    sample_data['column_b'],
    color="yellow",
    marker="x",
    markersize=8,
    markeredgecolor="black",
    markerfacecolor="yellow",
    label="B"
)

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```

```python
countries  = pd.read_csv("countries.csv")
```

```python
countries
```

```python
#We are now isolating data for the countries egypt and united states
EGY = countries[countries.country=="Egypt"]
USA = countries[countries.country=="United States"]
```

```python
plt.title("comparing population between EGY and USA")
plt.xlabel("population")
plt.ylabel("year")
plt.plot(EGY["year"] , EGY["population"]/10**6 , color = "red")
plt.plot(USA["year"] , USA["population"]/10**6, color = "blue")
plt.legend(["Egypt" , "United States"])
```

```python
plt.figure(figsize=(5,5), dpi = 100)
plt.title("comparing population growth between EGY and USA")
plt.xlabel("growth")
plt.ylabel("year")
plt.xticks([1950,1960,1970,1980,1990])
plt.yticks([100,150,200,250,300])
#diff way instead of plugging the labels into legend
plt.plot(EGY["year"] , EGY["population"]/EGY["population"].iloc[0]*100 , color = "#ff0505" , label = "Egypt" )
plt.plot(USA["year"] , USA["population"]/USA["population"].iloc[0]*100, color = "#0008ff" , label="United States")
plt.legend()
plt.savefig('growth.png' , dpi=300)
plt.show()
```

```python
fifa = pd.read_csv("fifa_data.csv")
```

```python
fifa
```

### Histograms

```python
# Define the bin boundaries for the histogram
bins = [40, 50, 60, 70, 80, 90, 100]

# Set the title and labels for the plot
plt.title("Overall Distribution")
plt.ylabel("Number of Players")
plt.xlabel("Overall")

# Create a histogram of the "Overall" column from the 'fifa' DataFrame
plt.hist(fifa["Overall"], bins=bins, color="#46474f")

# Display the histogram
plt.show()
```

```python
# Create subsets of player ratings for three different clubs
FCB = fifa[fifa["Club"]=="FC Barcelona"]["Overall"]
PSG = fifa[fifa.Club=="Paris Saint-Germain"]["Overall"]
CUD = fifa[fifa.Club=="Cambridge United"]["Overall"]
```

### Box plotting

```python
labels = ["FCB" , "PSG" , "CUD"]
plt.figure(figsize=(5,8))
plt.ylabel("Fifa overall" , color='b')
plt.xlabel("Clubs" , color = 'r')
plt.title("Fifa overall across teams")
plt.boxplot([FCB,PSG,CUD] ,labels = labels)
plt.show
#We can really see how balanced barcelona and PSG are against each other and it would be unfair to be matched with cambridge united
```
